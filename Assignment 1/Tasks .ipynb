```python
## Task 3a)

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

pi = np.pi
omega = pi
dt = 0.0001

t = np.arange(0,6,0.0001)

def z(N): 
    z = 0   
    for n in range(1, N+1):
        x = np.sin(n*omega*t)
        b = ((-1)**(n-1))*(2/(n*pi))
        z = z + x*b
        
    return z

z1 = z(1)
z2 = z(2)
z10 = z(10)
z20 = z(20)

        
# Create a new figure
fig, ax = plt.subplots()
        
#plot
ax.plot(t, z1, label='z1(t) (N=1)')
ax.plot(t, z2, label='z2(t) (N=2)')
ax.plot(t, z10, label='z3(t) (N=10)')
ax.plot(t, z20, label='z4(t) (N=20)')


ax.set_xlabel('t (s)')
ax.set_ylabel('z(t)')
ax.legend()

ax.grid()

ax.set_xlim(0,6)

fig.savefig('figure_task_3a')
```
