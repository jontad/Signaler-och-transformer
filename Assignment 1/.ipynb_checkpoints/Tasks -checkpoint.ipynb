```python
## Task 1

import numpy as np
import matplotlib. pyplot as plt
from scipy import signal

# Generate a time vector t
t = np.arange(0,6,0.0001)

# Generate two signals x1 and x2
T1 = 2
f2 = 4

pi = np.pi

x1 = np.sin((2*pi*t)/T1)
x2 = np.sin(2*pi*t*f2)

# Create a new figure
fig, ax = plt.subplots()

# plot
ax.plot(t , x1, label='x1(t)')
ax.plot(t , x2, label='x2(t)')

ax.set_xlabel('t (s)')
ax.set_ylabel('x(t)')
ax.legend()

ax.grid()

ax.set_xlim(0,6)
ax.set_ylim(-1.05,1.05)

fig.savefig('figure_task_1')

```

```python
## Task 2a)

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

pi = np.pi
alpha = 2*pi

t = np.arange(0,6,0.0001)
dt = 0.001

# Create impulse response
h = (alpha**2)*t*np.exp(-alpha*t)

# Create Dirac delta function
d = np.zeros(t.shape)
d[0] = 1/dt


y = dt*signal.convolve(d, h, method='direct')
y = y[0:d.shape[0]]


# Create a new figure
fig, ax = plt.subplots()

#plot
ax.plot(t, h, label='h(t)')
ax.plot(t, y, label='y(t)')

ax.set_xlabel('t (s)')
ax.set_ylabel('h(t), y(t)')
ax.legend()

ax.grid()

ax.set_xlim(0,3)
ax.set_ylim(0,2.5)

fig.savefig('figure_task_2a')

```

```python
## Task 2b)

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

pi = np.pi
dt = 0.001
t = np.arange(0,6,0.0001)

# Create impulse response
T1 = 2
f2 = 4

x1 = np.sin((2*pi*t)/T1)
x2 = np.sin(2*pi*t*f2)

# Create impulse response
alpha = 2*pi
h = (alpha**2)*t*np.exp(-alpha*t)

# Convolve Dirac delta function with x1(t) and x2(t)
y1 = dt*signal.convolve(x1, h, method='direct')
y2 = dt*signal.convolve(x2, h, method='direct')

y1 = y1[0:x1.shape[0]]
y2 = y2[0:x2.shape[0]]

# Create a new figure
fig, ax = plt.subplots()

#plot
ax.plot(t, y1, label='y1(t)')
ax.plot(t, y2, label='y2(t)')

ax.set_xlabel('t (s)')
ax.set_ylabel('y(t)')
ax.legend()

ax.grid()

ax.set_xlim(0,6)

fig.savefig('figure_task_2b')
```

```python
## Task 2c)

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

pi = np.pi
dt = 0.001
t = np.arange(0,6,0.0001)

# Create impulse response
T1 = 2
f2 = 4

x1 = np.sin((2*pi*t)/T1)
x2 = np.sin(2*pi*t*f2)

x3 = x1+x2

# Create impulse response
alpha = 2*pi
h = (alpha**2)*t*np.exp(-alpha*t)

# Convolve Dirac delta function with x1(t) and x2(t)
y1 = dt*signal.convolve(x1, h, method='direct')
y2 = dt*signal.convolve(x2, h, method='direct')
y3 = dt*signal.convolve(x3, h, method='direct')

y1 = y1[0:x1.shape[0]]
y2 = y2[0:x2.shape[0]]
y = y1+y2

y3 = y3[0:x2.shape[0]]

# Create a new figure
fig, ax = plt.subplots()

#plot
ax.plot(t, y, label='y1(t)+y2(t)')
ax.plot(t, y3, label='y3(t)')

ax.set_xlabel('t (s)')
ax.set_ylabel('y(t)')
ax.legend()

ax.grid()

ax.set_xlim(0,6)

fig.savefig('figure_task_2c')
```
