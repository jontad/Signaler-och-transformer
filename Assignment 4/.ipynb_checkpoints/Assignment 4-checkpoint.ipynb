```python
#Task 1
import numpy as np
import matplotlib. pyplot as plt
from scipy import signal

#constants
pi = np.pi
w1 = (21*pi)/64
w2 = (11*pi)/32
k = np.arange(0,20, 1)
print(k)

#discrete time cosine signals
x1 = np.cos(w1*k)
x2 = np.cos(w2*k)


#Plot
fig, ax = plt.subplots()
ax.stem(k, x1,  markerfmt='bo', linefmt='blue', label='$x_1[k]$')
ax.stem(k, x2,  markerfmt='ro', linefmt='red', label='$x_2[k]$')
ax.legend()


ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# remove the ticks from the top and right edges
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_ylabel('x[k]')
ax.set_xlabel('k')

ax.set_xticks(k)
plt.show()


fig.savefig('figure_task_1b')
```

```python
#Task 2
from scipy.fft import fft

#constants
fs = 128
K = 20
L = K

#frequency vector
fl = np.arange(0, L)/L*fs
fl = [i for i in fl if i <= fs/2]

#caluculate discrete fourier transform
X1 = fft(x1)
X2 = fft(x2)

#Plot only first halves
sliceObject = slice(len(fl))
X1 = X1[sliceObject]
X2 = X2[sliceObject]

#scale 
X1 = (1/K)*X1
X2 = (1/K)*X2

#magnitudes of DFT
X1_mag = np.abs(X1)
X2_mag = np.abs(X2)

#Plot
fig, ax = plt.subplots()
ax.stem(fl, X1_mag, markerfmt='bo', linefmt='blue', label='$|X_1(f_l)|$')
ax.stem(fl, X2_mag, markerfmt='ro', linefmt='red', label='$|X_2(f_l)|$')
ax.legend()

ax.set_ylabel('$|X(f_l)|$')
ax.set_xlabel('$f_l$')

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# remove the ticks from the top and right edges
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()

fig.savefig('figure_task_2a_mag')
    
```

```python
#Task 2
from scipy.fft import fft

#constants
fs = 128
K = 20
L = 256

#frequency vector
fl = np.arange(0, L)/L*fs



#caluculate discrete fourier transform
X1 = fft(x1, n=len(fl))
X2 = fft(x2, n=len(fl))

#Plot only first halves
fl = [i for i in fl if i <= fs/2]

sliceObject = slice(len(fl))
X1 = X1[sliceObject]
X2 = X2[sliceObject]

#scale 
X1 = (1/K)*X1
X2 = (1/K)*X2

#magnitudes of DFT
X1_padded_mag = np.abs(X1)
X2_padded_mag = np.abs(X2)

#Plot
fig, ax = plt.subplots()
ax.stem(fl, X1_padded_mag, markerfmt='bo', linefmt='blue', label='$|X_1(f_l)|$')
ax.stem(fl, X2_padded_mag, markerfmt='ro', linefmt='red', label='$|X_2(f_l)|$')
ax.legend()

ax.set_ylabel('$|X(f_l)|$')
ax.set_xlabel('$f_l$')

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# remove the ticks from the top and right edges
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.show()

    
fig.savefig('figure_task_2b_mag')

```
